```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import shutil
```

```python
with open('filenames.pkl', 'rb') as f:
    filenames = pickle.load(f)

df = pd.read_excel('data_extended.xlsx')
```

```python
df
```

```python
objects = ['Animal', 'Crown', 'Moon', 'Plant', 'Stars', 'Sun', 'Text', 'Tool', 'Weapon']
for object in objects:
    for i, filename in enumerate(filenames):
        if df[object][i] > 0:
            shutil.copyfile('svg_flags/' + filename + '.svg', 'svg_flags/Objects/' + object + '/' + filename  + '.svg')
```

```python
layouts = ['Circle', 'Cross', 'Saltires', 'Triangle']
for layout in layouts:
    for i, filename in enumerate(filenames):
        if df[layout][i] > 0:
            shutil.copyfile('svg_flags/' + filename  + '.svg', 'svg_flags/Layout/' + layout + '/' + filename  + '.svg')
```

```python
layouts = ['Bars', 'Stripes']
for layout in layouts:
    for i, filename in enumerate(filenames):
        if df[layout][i] > 0:
            shutil.copyfile('svg_flags/' + filename  + '.svg', 'svg_flags/Layout/' + layout + '/' + str(df[layout][i]) + '_' + filename  + '.svg')
```
